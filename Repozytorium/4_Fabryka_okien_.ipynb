```
from pulp import *
problem = LpProblem("Najlepszaprodukcja",LpMaximize)

x1 = LpVariable("godzinapartiI", 0, None, LpInteger)
x2 = LpVariable("godiznapartiII", 0, None, LpInteger)

problem += 3000.0*x1 + 5000.0*x2, "zysk"
problem += x1 + 0*x2 <= 4, "Fabryka1"
problem += 0*x1 + 2*x2 <= 12, "Fabryka2"
problem += 3*x1 + 2*x2 <= 18, "Fabryka3"


problem.solve()
print("Status:", LpStatus[problem.status])
```

```
for v in problem.variables():
    print (v.name, "=", v.varValue)
print ("Zysk: {} zl".format(value(problem.objective)))
```

```

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}


profit_aluminium_slider = FloatSlider(min=0,max = 3000, value= 1000, description="profit_aluminium", style= style )
profit_drewno_slider = FloatSlider(min=0,max = 5000,value= 1000, description="profit_drewno", style= style )
godz_fabryka1_slider = FloatSlider(min=0,max = 4, value= 2, description="dostepnosc_farbyki_pierwszej", style= style )
godz_fabryka2_slider = FloatSlider(min=0,max = 12, value= 2, description="dostepnosc_farbyki_drugiej", style= style )
godz_fabryka3_slider = FloatSlider(min=0,max = 18, value= 2, description="dostepnosc_farbyki_trzeciej", style= style )
ogra_al_1_slider = FloatSlider(min=0,max = 1, value= 1, description="ograniczenie_farbyki1_aluminium", style= style )
ogra_drew_1_slider = FloatSlider(min=0,max = 0, value= 0, description="ograniczenie_farbyki1_drewno", style= style )
ogra_al_2_slider = FloatSlider(min=0,max = 0, value= 0, description="ograniczenie_farbyki2_aluminium", style= style )
ogra_drew_2_slider = FloatSlider(min=0,max = 2, value= 0, description="ograniczenie_farbyki2_drewno", style= style )
ogra_al_3_slider = FloatSlider(min=0,max = 3, value= 1, description="ograniczenie_farbyki3_aluminium", style= style )
ogra_drew_3_slider = FloatSlider(min=0,max = 2, value= 0, description="ograniczenie_farbyki3_drewno", style= style )


def fabryka(
    profit_aluminium = 1000,
    profit_drewno = 1000,
    godz_fabryka1 = 2,
    godz_fabryka2 = 2,
    godz_fabryka3 = 2,
    ogra_al_fabryka1 = 1,
    ogra_drew_fabryka1 = 0,
    ogra_al_fabryka2 = 0,
    ogra_drew_fabryka2 = 0,
    ogra_al_fabryka3 = 1,    
    ogra_drew_fabryka3 = 0):
    
    problem = LpProblem("Rozkład",LpMaximize)
    x1 = LpVariable("Okna_aluminium", 0, None, LpInteger)
    x2 = LpVariable("Okna_drewniane", 0, None, LpInteger)
    problem += profit_aluminium*x1 + profit_drewno*x2, "Zysk"
    problem += ogra_al_fabryka1*x1 + ogra_drew_fabryka1*x2 <= godz_fabryka1, "F1"
    problem += ogra_al_fabryka2*x1 + ogra_drew_fabryka2*x2 <= godz_fabryka2, "F2"
    problem += ogra_al_fabryka3*x1 + ogra_drew_fabryka3*x2 <= godz_fabryka3, "F3" 

problem.solve()
print("Status:", LpStatus[problem.status])   

interact(fabryka,
    ogra_al_fabryka1 = ogra_al_1_slider,
    ogra_al_fabryka2 = ogra_al_2_slider,
    ogra_al_fabryka3 = ogra_al_3_slider,
    ogra_drew_fabryka1 = ogra_drew_1_slider,
    ogra_drew_fabryka2 = ogra_drew_2_slider,
    ogra_drew_fabryka3 = ogra_drew_3_slider,
    profit_aluminium = profit_aluminium_slider,
    profit_drewno =    profit_drewno_slider,
    ogarniczenie_fabryki1 =    godz_fabryka1_slider,
    ogarniczenie_fabryki2 =    godz_fabryka2_slider,
    ogarniczenie_fabryki3 =    godz_fabryka3_slider,
    )


```
